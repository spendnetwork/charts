```
import pandas as pd
import altair as alt
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta
```

```
url = 'https://raw.githubusercontent.com/spendnetwork/charts/master/ministry_of_defence_contract_data.csv'
df = pd.read_csv(url)
#print(df)
```

```
df['releasedate']=pd.to_datetime(df['releasedate']).dt.strftime('%m/%d/%Y')
#df.head()
```

```
def convert_nan_to_zero(val):
      if val == "NaN":
        return 0
      elif val == '':
        return 0
      else:
        return val
```

```
df['aw_value']=df["aw_value"].apply(convert_nan_to_zero)
#df.head()
```

```
def value_buckets(value): 
      value=int(value)
      if value < 500000: 
        value_group = ' less than 500K' 
      elif value < 1000000: 
        value_group = '500K-1M' 
      elif value < 2000000: 
        value_group = '1M-2M' 
      elif value < 3000000: 
        value_group = '2M-3M' 
      elif value < 4000000: 
        value_group ='3M-4M' 
      elif value >=5000000: 
        value_group = '4M-5M' 
      else: 
        value_group = 'more than 5M'
      return value_group
```

```
df['value_buckets'] = df["aw_value"].apply(value_buckets)
#df.head()
```

```
df = df.set_index(['releasedate','value_buckets'])
```

```
df2=(df.groupby(['releasedate','value_buckets']).agg({'ocid': ['count']}))
df2.head
```

```
df2.columns = df2.columns.map('_'.join)
df2 = df2.reset_index()
df2.head()
```

```
alt.Chart(df2,width=580, height=517, title='Spend by bands of Ministry of Defence').mark_line(strokeWidth=3).encode(
    alt.X('year(releasedate):O', axis=alt.Axis(tickCount=12, title="https://openopps.com", 
                grid=False, 
                labelFont='Helvetica', 
                labelFontSize=12, 
                labelAngle=0, 
                titleAlign='left',
                titleAnchor='start',
                titleFont='Helvetica',
                titleFontSize= 11,
                titleFontWeight= 300,
                titleX= -0,
                titleY= 30)), 
    alt.Y('sum(ocid_count):Q', axis=alt.Axis(title=None,
                                    format=".0s",
                                    titleAlign= 'left',
                                    titleAnchor='start',
                                    titleAngle= 0,
                                    titleFont= 'Helvetica',
                                    titleFontSize= 12,
                                    titleFontWeight= 400,
                                    titleX= -0,
                                    titleY= -30,)),
    alt.Color('value_buckets:O',title=None, scale=alt.Scale( range=[ "#3E9454",
                    "#006D8C", 
                    "#91D739", 
                    "#EC4256",
                    "#67E4C5", 
                    "#fDB6B4",])) ).configure_legend(
    gradientDirection ='vertical',
    orient='right'
).configure_view(
          strokeWidth=0
        ).configure_title(fontSize=18, anchor='start',
                color= "#91D739", font='Helvetica', fontWeight=400)
```
