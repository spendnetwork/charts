```
import pandas as pd
import altair as alt
```

```
url = 'https://raw.githubusercontent.com/spendnetwork/charts/master/minsitry_of_defence_tenders.csv'
df = pd.read_csv(url)
#print(df)
```

```
df['releasedate']=pd.to_datetime(df['releasedate']).dt.strftime('%m/%d/%Y')
```

```
df1=(df.groupby('releasedate').agg({'ocid': ['count'], 'value': 'sum'}))
#df1.head()
```

```
df1.columns = df1.columns.map('_'.join)
df1 = df1.reset_index()
#df1.head()
```

```
  nearest = alt.selection(type='single', nearest=True, on='mouseover',fields=['releasedate'], empty='none')
```

```
        line=alt.Chart(df1, width=1200, height=250, title="Count of tenders by Ministry of Defence").mark_line(color='#91D739').encode(
            x=alt.X('yearmonth(releasedate):O',
                    axis=alt.Axis(tickCount=12, title="https://openopps.com", 
                grid=False, 
                labelFont='Helvetica', 
                labelFontSize=12, 
                labelAngle=0, 
                titleAlign='left',
                titleAnchor='start',
                titleFont='Helvetica',
                titleFontSize= 11,
                titleFontWeight= 300,
                titleX= -0,
                titleY= 30) 
                   ),
            y=alt.Y('sum(ocid_count):Q',
                     axis=alt.Axis(title="",
                                    format=".0s",)
                   ),
            )
        selectors = alt.Chart(df1).mark_point().encode(
            x='yearmonth(releasedate):O',
            opacity=alt.value(0),
        ).add_selection(
            nearest
        )

        points = line.mark_point().encode(
            opacity=alt.condition(nearest, alt.value(1), alt.value(0))
        )


        text = line.mark_text(align='left', dx=10, dy=-10).encode(
            text=alt.condition(nearest, 'sum(ocid_count):Q', alt.value(' '))
        )

        rules = alt.Chart(df1).mark_rule(color='gray').encode(
            x='yearmonth(releasedate):O',
        ).transform_filter(
            nearest
        )

        chart = alt.layer(
            line, selectors, points, rules, text
        ).configure_view(
          strokeWidth=0
        ).configure_title(fontSize=18, anchor='start',
                color= "#91D739", font='Helvetica', fontWeight=400)

        chart
```
